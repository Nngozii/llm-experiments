#### Use of LangChain (powerful, mighty and heavy) and LiteLLM (lightweight, efficient)

```python
# LangChain

from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model= "gpt-5-nano")
response=llm.invoke(message)

display(Markdown(response.content))
```

```python
# LiteLLM

from litellm import completion

response = completion(
    model = "gpt-5-nano",
    messages = message
)
display(Markdown(response.choices[0].message.content))
```

With LiteLLM, we can check number of tokens

```python
input_tokens = response.usage.prompt_tokens
output_tokens = response.usage.completion_tokens
total_tokens = response.usage.total_tokens

print(f"Input tokens: {input_tokens}")
print(f"Output tokens: {output_tokens}")
print(f"Total tokens: {total_tokens}")
```

### Game Time

A fun and adversarial conversation between chatbots...

```python
# Between 2 Models
openai = OpenAI()
deepseekk = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")

gpt_model = "gpt-5-nano"
deepseek_model = "deepseek-r1:1.5b"

gpt_system_prompt = """
You are a chatbot who is very argumentative; You disagree with\
anything in the conversation and challenge everything in a snarky way.
"""

deepseek_system_prompt = """
You are a chatbot who is very sarcastic; You respond every message from\
the other person in a calm but sarcastic way. 

"""

gpt_message = ["Hi, there"]
deepseek_message = ["Hi"]
```

```python
# gpt function

def call_gpt():
    message = [{"role":"system", "content": gpt_system_prompt}]
    #zip() pairs items from two lists at the same position.
    for gpt, deepseek in zip(gpt_message, deepseek_message):
        message.append({"role":"assistant", "content":gpt})
        message.append({"role":"user", "content":deepseek})
    response = openai.chat.completions.create(
        model = gpt_model,
        messages= message
    )
    return response.choices[0].message.content
```

```python
call_gpt()
```

```python
# deepseek function

def call_deepseek():
    message = [{"role": "system", "content": deepseek_system_prompt}]
    for gpt, deepseek in zip(gpt_message, deepseek_message):
        message.append({"role":"user", "content":gpt})
        message.append({"role":"assistant", "content":deepseek})
    message.append({"role":"user", "content":gpt_message[-1]})
    response = deepseekk.chat.completions.create(
        model = deepseek_model,
        messages= message
    )
    return response.choices[0].message.content
```

```python
call_gpt()
```

```python
call_deepseek()
```

```python
display(Markdown(f"### GPT:\n{gpt_message[0]}\n"))
display(Markdown(f"### Deepseek:\n{deepseek_message[0]}\n"))

for i in range (3):
    gpt_next = call_gpt()
    display(Markdown(f"### GPT:\n{gpt_next}"))
    gpt_message.append(gpt_next)
    
    deepseek_next = call_deepseek
    display(Markdown(f"### Deepseek:\n{deepseek_next}"))
    deepseek_message.append(deepseek_next)
```
