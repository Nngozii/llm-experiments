```python
# Importing packages

import os
from IPython.display import Markdown, display, update_display
from openai import OpenAI
from dotenv import load_dotenv
```

```python
load_dotenv(override=True)
MISTRAL_BASE_URL = "https://api.mistral.ai/v1"
OLLAMA_BASE_URL = "http://localhost:11434/v1"
GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"

mistral_api_key = os.getenv('MISTRAL_API_KEY')
openai_api_key = os.getenv('OPENAI_API_KEY')
gemini_api_key = os.getenv('GOOGLE_API_KEY')

if mistral_api_key:
    print(f"Key found and begins with {mistral_api_key[:5]}")
else:
    print("key not found")
if openai_api_key:
    print(f"Key found and begins with {openai_api_key[:5]}")
else:
    print("key not found")
if gemini_api_key:
    print(f"Key found and begins with {gemini_api_key[:5]}")
else:
    print("key not found")
```

```python
mistral = OpenAI(base_url=MISTRAL_BASE_URL, api_key=mistral_api_key)
gemini = OpenAI(base_url=GEMINI_BASE_URL, api_key=gemini_api_key)
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")
```

```python
message = [{"role": "user", "content":"tell a joke"}]
```

```python
# GPT
client = OpenAI()

response = client.chat.completions.create(
    model = "gpt-5-nano",
    messages=message
)

display(Markdown(response.choices[0].message.content))
```

```python
# Mistral

response = mistral.chat.completions.create(
    model = "mistral-small-latest",
    messages=message
)

display(Markdown(response.choices[0].message.content))

# So um, apparently, Mistral doesn't have small model I can run om my PC
```

```python
# Gemini

response = gemini.chat.completions.create(
    model = "gemini-2.5-flash",
    messages=message
)

display(Markdown(response.choices[0].message.content))
```
